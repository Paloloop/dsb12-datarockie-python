```python
## map/fitler -> iterable (adj) จะได้ไม่ต้องใช้ for-loop
scores = [82, 78, 81, 66, 50]
print(scores)

## map
# lamda function - add five points to scores
scores = list(map(lambda x: x+5, scores))
print(scores)

def direct_grading(score):
  if score >= 80:
    return "Passed"
  else:
    return "Not Pass"

score_result = list(map(direct_grading, scores))
print(score_result)

print("-------------------------------------------")
## filter
pass_scores = list(filter(lambda x:x>=80, scores))
print(pass_scores)
```

```python
#Restaurant Chatbot
def chat():
  print("Good morning sir!")
  order_list = {"pizza" : 0, "hotdog" : 0, "cola" : 0}
  price = {"pizza" : 50, "hotdog" : 20, "cola" :15}
  total_bill = 0

  order = input("What do you want to order : ").strip().lower()
  while order not in order_list:
    print("We dont have that one. We have pizza, hotdog, and cola")
    order = input("What do you want to order : ").strip().lower()
  order_list[order] += 1
  print(f"{order} has been added to your order.")

  while True:
    next_order = input("Anything else? : ").strip().lower()

    while (next_order not in order_list) and next_order != "no":
      print("We dont have that one. We have pizza, hotdog, and cola")
      next_order = input("Anything else? : ").strip().lower()
    # end the process
    if next_order == "no":
      # calculate total bill
      for item in order_list:
        total_bill += order_list[item]*price[item]
      print(f"Your Total Amount is {total_bill}")
      break
    order_list[next_order] += 1
    print(f"{next_order} has been added to your order.")
chat()
```

```python
class Dog:
  name = "Toy"
  age = 5
  color = "Brown"
  breed = "French Bulldog"

  #function (Dog method)
  def sitting(self):
    print("I am sitting now!")

  def hungry(self, food_name):
    print(f"I am hungry, I need {food_name}!")
```

```python
# ATM - deposit, withdraw, chack balance, OTP, order coffee, pay bill, internet etc.
class ATM:
  def __init__(self, username, bank, amount):
    self.username = username
    self.bank = bank
    self.amount = amount

  def __str__(self):
    return f"{self.username} uses {self.bank} and has {self.amount} THB in the account"

  def check_balance(self):
    print(f"Balance: {self.amount} THB")

  def deposit(self, saving):
    self.amount += saving
    print(f"Deposit {saving} THB successfully")
```

```python
peach = ATM("Palapol Peach", "kbank", 100)
peach.check_balance()
peach.deposit(100)
peach.check_balance()
print(peach)
```

```python
## super class
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

# extension
class Employee(Person):
  def __init__(self, name, age, company):
    super().__init__(name, age) #super จะ refer ไปหา Person
    self.company = company

  def greeting(self):
    print(f"hello {self.name} from {self.company}")
```

```python
person1 = Person("toy", 37)
```

```python
person2 = Employee("lisa", 28, "google")
person2.greeting()
```

### Date Format : `.dt.strftime('%b-%Y')`
- return String
---

*   **`.dt`**: This is the accessor for datetime properties of the Series.
*   **`.strftime()`**: This method formats datetime objects into string representations based on a format code.
*   **`'%b'`**: This format code represents the abbreviated month name (e.g., 'Jan', 'Feb', 'Mar').
*   **`'%Y'`**: This format code represents the year with century as a decimal number (e.g., '2019', '2020').

You can experiment with different format codes to get other date and time representations. For example, `'%d-%m-%Y'` would give you '01-11-2019' for November 8, 2019.

### Group by
