### TODO 05 - how much total sales, average sales, and standard deviation of sales your company make in 2017

```python
salein2017 = data.query('`Order Date`.dt.year == 2017')
salein2017.head()
summary = salein2017.describe()
display(summary)
total_sale = salein2017['Sales'].sum()
avg_sale = salein2017['Sales'].mean()
#std_sale = summary['Sales']['std'] อย่างนี้จะ index สองครั้ง
std_Sale = summary.loc['std', 'Sales'] #index รวดเดียวเลย summary.loc['std', 'Sales']
print(f'Total Sales : {total_sale}, Average Sales : {avg_sale}, STD Sales : {std_Sale}')
```

### TODO 08 - what is the proportion of total sales (%) in West + Central in 2019 e.g. 25%

```python
salesin2019 = data.query('`Order Date`.dt.year == 2019')
display(salesin2019.head())

total_sale = salesin2019['Sales'].sum()
print(f"Total Sale : {total_sale}")

#westandcentral = salesin2019[(salesin2019['Region'] == "West") | (salesin2019['Region'] == "Central")]['Sales'].sum() #multiple-condition
westandcentral = salesin2019[salesin2019['Region'].isin(["West", "Central"])]['Sales'].sum() #รวม value เหมือน IN ('West','Central')
print(f"West + Central Sales : {westandcentral}")
prop_sales = westandcentral / total_sale
print(f"Proportion of Total Sales : {prop_sales}")
print(f"Proportion of Total Sales : {prop_sales:.2%}")
```

### TODO 09 - find top 10 popular products in terms of number of orders vs. total sales during 2019-2020

```python
# Using Merge Method
salefrom19to20 = data.query('(`Order Date`.dt.year).isin([2019,2020])')
print("salefrom19to20")
display(salefrom19to20.head())

#top product count- table
top10product = salefrom19to20[['Product ID','Product Name']].value_counts().reset_index()
print("Top 10 Product")
display(top10product.head())

#sales by product - table
agg_sales = salefrom19to20.groupby('Product ID')['Sales'].sum()
print("Total Sale by Product")
display(agg_sales.head())

#top product count vs sales - table
final_df = pd.merge(top10product,agg_sales,on="Product ID").sort_values(['count','Sales'],ascending = False)
print("Final Result")
display(final_df.head(10))
```

```python
# Using Multiple Group By
salefrom19to20 = data.query('(`Order Date`.dt.year).isin([2019,2020])')
print("salefrom19to20")
display(salefrom19to20.head())

final = salefrom19to20.groupby(['Product ID', 'Product Name'])['Sales']\
          .agg(['size','sum'])\
          .sort_values(['size','sum'], ascending=False)\
          .reset_index()
display(final.head(10))

net_profit_final = salefrom19to20.groupby(['Product ID', 'Product Name'])\
                    .agg(order_count = ('Product ID','size'),
                         total_sales = ('Sales','sum'),
                         net_profit = ('Profit','sum'))\
                    .sort_values(['order_count','total_sales','net_profit'], ascending=False)\
                    .reset_index()
display(net_profit_final.head(10))
#เริ่ม column ด้วย product id กับ product name
#add column ที่เป็น count product id เข้าไปใน agg
#add column ที่เป็น sum sales เข้าไปใน agg
#add column ที่เป็น sum profit เข้าไปใน agg
```

```python
#Pivot table
salefrom19to20 = data.query('(`Order Date`.dt.year).isin([2019,2020])')
print("salefrom19to20")
display(salefrom19to20.head())

final = salefrom19to20.pivot_table(
    index=['Product ID', 'Product Name'],\
    values='Sales',\
    aggfunc=['count', 'sum']
)

# Flatten column names
final.columns = ['order_count', 'total_sales']
final = final.sort_values(
    ['order_count', 'total_sales'],
    ascending=False
).reset_index()
display(final.head(10))
```

### TODO 10 - plot at least 2 plots, any plot you think interesting

```python
import matplotlib.pyplot as plt
import pandas as pd

## Plot 1 - Sales by Date
# --- Step 1: Prepare data (correct time handling) ---
sales_by_year = data.groupby(data['Order Date'].dt.to_period('M'))[['Sales', 'Profit']] \
                    .sum() \
                    .sort_index() \
                    .reset_index()

# --- Step 2: Create readable label ---
sales_by_year['Month-Year'] = sales_by_year['Order Date'].dt.strftime('%b-%Y')
display(sales_by_year.head())

# --- Step 3: Plot ---
fig, ax1 = plt.subplots(figsize=(12, 6))

# Left axis → Total Sales
line1, = ax1.plot(sales_by_year['Month-Year'],
                  sales_by_year['Sales'],
                  color='salmon',
                  label='Total Sales') #symbol box

ax1.set_xlabel('Month-Year')
ax1.set_ylabel('Total Sales')
ax1.tick_params(axis='x', rotation=45)

# Right axis → Net Profit
ax2 = ax1.twinx()

line2, = ax2.plot(sales_by_year['Month-Year'],
                  sales_by_year['Profit'],
                  color='green',
                  label='Net Profit')
ax2.set_ylabel('Net Profit')
profit_max = sales_by_year['Profit'].max()
ax2.set_ylim(0, profit_max * 2)

# --- Step 4: Combined legend (important) ---
lines = [line1, line2]
labels = [l.get_label() for l in lines]
ax1.legend(lines, labels, loc='upper left')

# --- Step 5: Final touches ---
plt.title('Total Sales and Net Profit by Month-Year')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

## Plot 1 - Sales by Date
# --- Step 1: Prepare data (correct time handling) ---
sales_by_year = data.groupby(data['Order Date'].dt.to_period('M'))[['Sales', 'Profit']] \
                    .sum() \
                    .sort_index() \
                    .reset_index()

# --- Step 2: Create readable label ---
sales_by_year['Month-Year'] = sales_by_year['Order Date'].dt.strftime('%b-%Y')
display(sales_by_year.head())

fig, ax1 = plt.subplots(figsize=(12, 6))

# Line (Sales)
line1, = ax1.plot(
    sales_by_year['Month-Year'],
    sales_by_year['Sales'],
    color='salmon',
    label='Total Sales'
)

ax1.set_xlabel('Month-Year')
ax1.set_ylabel('Total Sales')
ax1.tick_params(axis='x', rotation=45)

# Bar (Profit)
ax2 = ax1.twinx()

bars = ax2.bar(
    sales_by_year['Month-Year'],
    sales_by_year['Profit'],
    color='green',
    alpha=0.3,
    label='Net Profit'
)

ax2.set_ylabel('Net Profit')

# Better scaling
profit_max = sales_by_year['Profit'].max()
ax2.set_ylim(0, profit_max * 2)

# Legend fix
lines = [line1, bars]
labels = ['Total Sales', 'Net Profit']
ax1.legend(lines, labels, loc='upper left')

# Grid fix
ax1.grid(True, axis='y', linestyle='--', alpha=0.5)

plt.title('Total Sales and Net Profit by Month-Year')
plt.tight_layout()
plt.show()
```
